```python
import os
import csv
import requests
from http import HTTPStatus
from lxml import html
from bs4 import BeautifulSoup

url = "https://gz.lianjia.com/zufang/"
web = requests.session()

headers = {'Cookie': 'lianjia_ssid=1f27de47-a1a1-46fd-b561-50f4f72b4eb1; expires=Sun, 08-Sep-19 09:00:46 GMT; Max-Age=1800; domain=.lianjia.com; path=/',
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.131 Safari/537.36'
          }
zone = 'tianhe'
price = 'rp2'
r = web.get("{}/{}/{}".format(url, zone, price), headers=headers)
print(r)
```

```python
print(r.text)
```

```python
html = r.text
soup = BeautifulSoup(html, 'lxml')
ret_list = []
print(soup)
```

```python
temp_list = soup.find_all('div',class_='content__list--item')
print(temp_list)
```

```python
print(temp_list[0])
```

```python
temp_main_list = temp_list[0].find_all('div',class_='content__list--item--main')
print(temp_main_list)
```

```python
print(temp_main_list[0].find('p',class_='content__list--item--des').get_text(strip=True))
```

```python
print(temp_main_list[0].find('p',class_='content__list--item--des').get_text(strip=True).split('/'))
```

```python
for i in temp_main_list[0].find('p',class_='content__list--item--des').get_text(strip=True).split('/'):
    print(i.replace(' ',''))
```

```python
temp_price = temp_list[0].find_all(class_='content__list--item-price')
print(temp_price)
```

```python
temp_price1 = temp_list[0].em.get_text()
print(temp_price1)
```

```python
temp_price2 = temp_list[0].find(class_='content__list--item-price').em.get_text()
print(temp_price2)
```

```python
for i in soup.find_all('div',class_='content__list--item'):
    for j in i.find_all('div', class_= 'content__list--item--main'):
        des = j.find('p',class_='content__list--item--des').get_text(strip=True)
        time = j.find('p', class_ = 'content__list--item--time oneline').get_text(strip=True)
        des_item = des.split('/')
        location = des_item[0]
        area = des_item[1]
        toward = des_item[2]
        house_type = des_item[3]
        floor = des_item[4].replace(' ','')
        price = j.find(class_='content__list--item-price').em.get_text(strip=True)
        house =[location,area,toward,house_type,floor, time, price]
        ret_list.append(house)
print(ret_list)
```

```python
with open('lianjia.csv', 'w', encoding='utf-8') as f:
    writer_f = csv.writer(f)
    writer_f.writerow(['位置', '面积', '朝向', '户型', '楼层',  '发布时间','价格'])
    writer_f.writerows(ret_list)
```
