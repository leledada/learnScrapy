```python
 def save_csv(rows):
        headers = ["招聘公司","招聘职位","薪资","学历要求","职位详细页面url","职位详细内容"]
        os.makedirs('./zhaopin/', exist_ok=True)
        with open("./zhaopin/zhaopin.csv", "w", encoding="utf-8") as f: 
            f_csv = csv.writer(f)
            f_csv.writerow(headers)
            f_csv.writerows(rows)
```

```python
job_list = r.json()["data"]["results"]
print(len(job_list))
```

```python
rows = []
for n in range(len(job_list)):
    recruitment_company = job_list[n]["company"]["name"]
    recruitment_position = job_list[n]["jobName"]
    salary = job_list[n]["salary"]
    edu_level = job_list[n]["eduLevel"]["name"]
    detailed_page_url = job_list[n]["positionURL"]
    detailed_page_content = ''
    info_tuple = (
            recruitment_company, recruitment_position, salary, edu_level,
                detailed_page_url, detailed_page_content
                )
    rows.append(info_tuple)
```

```python
print(rows)
```

```python
save_csv(rows)
```
